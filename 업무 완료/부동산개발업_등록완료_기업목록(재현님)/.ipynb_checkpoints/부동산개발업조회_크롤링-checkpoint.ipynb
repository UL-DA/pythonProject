```python
# 한 페이지 테스트

# 페이지 오픈
driver = webdriver.Chrome('./driver/chromedriver.exe')
driver.get('http://www.nsdi.go.kr/lxportal/?menuno=4083') # 페이지 열기

# 검색
selectbox = Select(driver.find_element_by_id('shRegStsCd')) # 셀렉트박스 선택
selectbox_select = selectbox.select_by_index(1) # 등록완료 선택
selectbox_click  = driver.find_element_by_xpath('//*[@id="searchVO"]/div[3]/div/div[2]').click() # 검색 버튼 클릭

table_df = pd.DataFrame()
# 한 페이지의 기업 가져오기
for xpath_index in tqdm(range(2)): #9
    xpath_index = xpath_index + 1
    corp_click  = driver.find_element_by_xpath(f'//*[@id="searchVO"]/div[3]/table/tbody/tr[{xpath_index}]/td[2]/font').click() # 기업 클릭
    
    url = driver.current_url
    req = requests.get(url)
    df  = pd.read_html(req.text)[0]
    df_index = pd.concat([df[0], df[2], df[4]],axis = 0)
    df_val   = pd.concat([df[1], df[3], df[5]],axis = 0)
    
    table = pd.DataFrame({'index' : df_index, 'value' : df_val})
    index_list  = list(range(len(table)))
    table.index = index_list
    table = table.drop([6,12,15], axis=0, inplace=False)
    table = table.transpose()
    table_df = table_df.append(table)
    
    driver.back() # 뒤로가기
```

```python
# 중복제거
table_df.drop_duplicates()
```

```python
# 크롬 버전 : 103.0.5060.53
driver = webdriver.Chrome('./driver/chromedriver.exe')
driver.get('http://www.nsdi.go.kr/lxportal/?menuno=4083')

table_df = pd.DataFrame()

# 검색
selectbox = Select(driver.find_element_by_id('shRegStsCd'))
selectbox_select = selectbox.select_by_index(1)
selectbox_click  = driver.find_element_by_xpath('//*[@id="searchVO"]/div[3]/div/div[2]').click()

# 다음 페이지 (272p 까지)
for page in tqdm(range(271)):
    # 한 페이지 기업 정보 가져오기 
    for xpath_index in tqdm(range(10)):
        xpath_index = xpath_index + 1
        corp_click  = driver.find_element_by_xpath(f'//*[@id="searchVO"]/div[3]/table/tbody/tr[{xpath_index}]/td[2]/font').click()

        url = driver.current_url
        req = requests.get(url)
        df  = pd.read_html(req.text)[0]
        df_index = pd.concat([df[0], df[2], df[4]],axis = 0)
        df_val   = pd.concat([df[1], df[3], df[5]],axis = 0)

        table = pd.DataFrame({'index' : df_index, 'value' : df_val})
        index_list  = list(range(len(table)))
        table.index = index_list
        table = table.drop([6,12,15], axis=0, inplace=False)
        table = table.transpose()
        table_df = table_df.append(table)

        driver.back()
    
    # 다음 페이지 이동
    nextPage_click  = driver.find_element_by_xpath('//*[@id="searchVO"]/div[3]/div[2]/ol/li[12]/a').click()

driver.close()
table_df_ = table_df.drop_duplicates()
```

```python
# 272p 마지막 5개 기업 정보 가져오기
driver = webdriver.Chrome('./driver/chromedriver.exe')
driver.get('http://www.nsdi.go.kr/lxportal/?menuno=4083')

table_df = pd.DataFrame()

# 검색
selectbox = Select(driver.find_element_by_id('shRegStsCd'))
selectbox_select = selectbox.select_by_index(1)
selectbox_click  = driver.find_element_by_xpath('//*[@id="searchVO"]/div[3]/div/div[2]').click()
```

```python
# 5개 기업 정보 가져오기 
for xpath_index in tqdm(range(5)):
    xpath_index = xpath_index + 1
    corp_click  = driver.find_element_by_xpath(f'//*[@id="searchVO"]/div[3]/table/tbody/tr[{xpath_index}]/td[2]/font').click()

    url = driver.current_url
    req = requests.get(url)
    df  = pd.read_html(req.text)[0]
    df_index = pd.concat([df[0], df[2], df[4]],axis = 0)
    df_val   = pd.concat([df[1], df[3], df[5]],axis = 0)

    table = pd.DataFrame({'index' : df_index, 'value' : df_val})
    index_list  = list(range(len(table)))
    table.index = index_list
    table = table.drop([6,12,15], axis=0, inplace=False)
    table = table.transpose()
    table_df = table_df.append(table)

    driver.back()
    
driver.close()
table_df_1 = table_df.drop_duplicates()
```
