```python
# 버전1 -> 목록 이동 엑스패스가 다른 페이지들이 있음 -> 에러남
driver = webdriver.Chrome('./driver/chromedriver.exe')
driver.get('https://cu.bgfretail.com/store/list.do?category=store')

corp_list = [] # 매장명
addr_list = [] # 주소

# '점'으로 지점명 검색
searchByName = driver.find_element_by_xpath('//*[@id="jumpoName"]')
searchByName.clear()
searchByName.send_keys('점')
click = driver.find_element_by_xpath('//*[@id="contents"]/div[1]/div/div[1]/div[2]/table/tbody/tr[2]/td/a').click()

time.sleep(1)

for page in tqdm(range(3)):
    for corp_index in range(1,6):
        corp = driver.find_element_by_xpath(f'//*[@id="result_search"]/div[2]/div[1]/table/tbody/tr[{corp_index}]/td[1]/span[1]').text
        addr = driver.find_element_by_xpath(f'//*[@id="result_search"]/div[2]/div[1]/table/tbody/tr[{corp_index}]/td[2]/div/address/a').text
        
        time.sleep(1)
        
        corp_list.append(corp)
        addr_list.append(addr)
        
    # 다음 목록 이동
    paging = driver.find_element_by_xpath(f'//*[@id="paging"]/a').get_attribute('text')
    if paging == '다음':
        nextPage = driver.find_element_by_xpath(f'//*[@id="paging"]/a').click()
    else :
        pass

    time.sleep(1)

driver.close()
```

```python
def selectBox(XPath):
    sido_names = []
    
    sido_selectbox_list = driver.find_element_by_xpath(XPath) # 셀렉트 박스 선택
    sido_list  = sido_selectbox_list.find_elements_by_tag_name('option')  # 셀렉트 박스 안에 있는 옵션 리스트 가져오기
    sido_names = [option.get_attribute('value') for option in sido_list]  # 가져온 옵션을 리스트에 담기
    sido_names.remove('') # 첫째 칸에 있는 공백 제거
    print(sido_names)
```

```python
sido_names   = []
Gugun_names  = []
Dong_names   = []
Gugun_names_ = []
Dong_names_  = []

# 1. 시도 정보 추출
sido_selectbox_list = driver.find_element_by_xpath('//*[@id="sido"]') # 셀렉트 박스 선택
sido_list  = sido_selectbox_list.find_elements_by_tag_name('option')  # 셀렉트 박스 안에 있는 옵션 리스트 가져오기
sido_names = [option.get_attribute('value') for option in sido_list]  # 가져온 옵션을 리스트에 담기
sido_names.remove('') # 첫째 칸에 있는 공백 제거

for sido in tqdm(sido_names):
    # 1-1. 셀렉트 박스에서 선택되도록 처리
    sido_element = driver.find_element_by_id('sido')
    sido_element.send_keys(sido)
    
    time.sleep(1)
    
    # 2. 군구 정보 추출
    Gugun_selectbox_list = driver.find_element_by_xpath('//*[@id="Gugun"]') # 셀렉트 박스 선택
    Gugun_list = Gugun_selectbox_list.find_elements_by_tag_name('option')   # 셀렉트 박스 안에 있는 옵션 리스트 가져오기
    Gugun_names_ = [option.get_attribute('value') for option in Gugun_list] # 가져온 옵션을 리스트에 담기
    Gugun_names_.remove('') # 첫째 칸에 있는 공백 제거
    Gugun_names.extend(Gugun_names_)
    
    time.sleep(1)
    
    for Gugun in Gugun_names:
        # 2-1. 셀렉트 박스에서 선택되도록 처리
        Gugun_element = driver.find_element_by_id('Gugun')
        Gugun_element.send_keys(Gugun)

        time.sleep(1)

        # 3. 동 정보 추출
        Dong_selectbox_list = driver.find_element_by_xpath('//*[@id="Dong"]') # 셀렉트 박스 선택
        Dong_list = Dong_selectbox_list.find_elements_by_tag_name('option')   # 셀렉트 박스 안에 있는 옵션 리스트 가져오기
        Dong_names_ = [option.get_attribute('value') for option in Dong_list] # 가져온 옵션을 리스트에 담기
        Dong_names_.remove('') # 첫째 칸에 있는 공백 제거
        Dong_names.extend(Dong_names_)

        time.sleep(1)
```
