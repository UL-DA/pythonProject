```python
# 첫번째 기업이 반복된다. -> 셀레니움이 너무 빨리 넘어가서 그럼
# 방법1. for문의 마지막에 time.sleep(1) 주면 해결됨
```

```python
driver = webdriver.Chrome('./driver/chromedriver.exe')
driver.get('http://gs25.gsretail.com/gscvs/ko/store-services/locations#;')

corp_list = [] # 매장명
addr_list = [] # 주소

for page in tqdm(range(3151)):
    for corp_index in range(1,6):
        corp = driver.find_element_by_xpath(f'//*[@id="storeInfoList"]/tr[{corp_index}]/td[1]/a').get_attribute('text')
        addr = driver.find_element_by_xpath(f'//*[@id="storeInfoList"]/tr[{corp_index}]/td[2]/a').get_attribute('text')
        
        time.sleep(1)
        corp_list.append(corp)
        addr_list.append(addr)
        
    driver.find_element_by_xpath('//*[@id="pagingTagBox"]/a[3]').click()
    time.sleep(1)

driver.close()
```
