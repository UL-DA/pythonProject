```python
base = pd.read_csv('./법정동코드.csv')
```

```python
base['법정동코드'] = base['법정동코드'].astype(str)
```

```python
base = base[(~base['법정동코드'].str.contains('00000'))&(base['법정동코드'].str.endswith('00'))]
```

```python
base
```

```python
from tqdm import tqdm
DF = pd.DataFrame()
for i in tqdm(range(5055)):
    DF_ = df[(df['상세주소'].str.contains(list(base['법정동명'])[i].split(' ')[-1]))&(df['상세주소'].str.contains(list(base['법정동명'])[i].split(' ')[-2]))]
    DF_['동코드'] = [list(base['법정동코드'])[i]]*len(DF_)
    DF = DF.append(DF_)
    
```
