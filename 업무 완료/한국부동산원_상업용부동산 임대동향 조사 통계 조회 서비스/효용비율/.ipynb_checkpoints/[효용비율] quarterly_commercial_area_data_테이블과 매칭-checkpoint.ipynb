```python
df_202202 = df[df['research_date'] == 202202]
```

```python
df_202202 = df_202202.reset_index(drop=True)
```

```python
# 년, 분기로 나누고, 기존 날짜 컬럼 삭제
df['base_year'] = df['research_date'].astype(str).str[:4].astype(int)
df['base_quarter'] = df['research_date'].astype(str).str[-1:]
df['base_year'] = df['base_year'].astype(int)
df['base_quarter'] = df['base_quarter'].astype(int)

df = df.drop(columns=['research_date'])
```
